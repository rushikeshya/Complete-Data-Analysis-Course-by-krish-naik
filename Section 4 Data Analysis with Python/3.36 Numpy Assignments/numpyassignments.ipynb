### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
## 3.1 Create two NumPy arrays of shape (3,4) filled with random integers, Perform element-wise addition, subtraction, multiplication, and division
arr1 = np.random.randint(1,21, size=(3,4))
print("Array 1 = ")
print(arr1)
arr2 = np.random.randint(1,21, size=(3,4))
print("Array 2 = ")
print(arr2)

print("Element wise Addition: ")
addition  = arr1 + arr2
print(addition)

print("Element wise Substraction: ")
addition  = arr1 - arr2
print(addition)

print("Element wise Multiplication: ")
addition  = arr1 * arr2
print(addition)

print("Element wise Divition: ")
addition  = arr1 / arr2
print(addition)

```

```python
## 3.2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.
arr = np.arange(1,17).reshape(4,4)
print("Origin Array: ")
print(arr)

row_wise_sum = np.sum(arr,axis=1)
print(f"Row wise addition: {row_wise_sum}")

column_wise_sum = np.sum(arr, axis=0)
print(f"Column wise addition: {column_wise_sum}")
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
## 4.1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
arr = np.random.randint(1,21, size=(5,5))
print("Origin Array: ")
print(arr)

mean = np.mean(arr)
print(f"Mean = {mean}")

median = np.median(arr)
print(f"Median = {median}")

std = np.std(arr)
print(f"Standard Deviation = {std :.2f}")

varience = np.var(arr)
print(f"Varience = {varience :.2f}")
```

```python
## 4.2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).
arr = np.arange(1,10).reshape(3,3)
print("Original Array:")
print(arr)

mean = np.mean(arr)
std = np.std(arr)
normalized_array = (arr - mean) /std
print("Normalized Array: ")
print(normalized_array)
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python
## 6.1 
matrix = np.random.randint(1,21,size=(3,3))
print("Matrix: ")
print(matrix)

determinant = np.linalg.det(matrix)
print(f"determinant: {determinant}")

inverse = np.linalg.inv(matrix)
print(inverse)

eigenvalues = np.linalg.eigvals(matrix)
print(eigenvalues)
```

```python
## 6.2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.
arr1 = np.random.randint(1,11,size=(2,3))
arr2 = np.random.randint(1,11,size=(3,2))

print("Array 1: ")
print(arr1)

print("Array 2: ")
print(arr2)

print("Matrix Multiplication using `dot()` function")
matrix_multi = np.dot(arr1, arr2)
print(matrix_multi)
```
