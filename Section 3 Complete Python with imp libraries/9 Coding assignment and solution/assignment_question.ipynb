**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
num = int(input("Enter a number: "))

factorial = 1
if num < 0:
    print("Factorial dose not exist for negative numbers.")
else:
    for i in range(1, num+1):
        factorial = factorial * i
    print("Factorial of", num, "is:", factorial)

```

```python
## Using while loop

num = int(input("Enter a number: "))
factorial = 1
i = 1
if num < 0:
    print("Factorial does not exist for Negatives numbers.")
else:
    while i <= num:
        factorial = factorial * i
        i += 1
    print(f"Factorial of {num} is: {factorial}")
```

```python
## Factorial using Recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
num = int(input("Enter a number: "))
print(f"Factorial of {num} is: {factorial(num)}")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input("Enter the Year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a LEAP Year")
else:
    print(f"{year} is a NOT LEAP Year")
```

**Question 18:** Write a Python program to reverse a string.

```python
# Your code here
string = input("Enter a string: ")

reversed_string = ""

for char in string:
    reversed_string = char + reversed_string

print("Reversed string: ", reversed_string)
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
string = input("Enter the string")

if string == string[::-1]:
    print(f"{string} is palindrome.")
else:
    print(f"{string} is not a palindrome.")
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Your code here
n = int(input("Enter the Lenght of list: "))
lst = []

for i in range(n):
    element = int(input(f"Enter the element of the list {i + 1}: "))
    lst.append(element)

lst.sort()
print("Sorted list in ascending order:", lst)
```
