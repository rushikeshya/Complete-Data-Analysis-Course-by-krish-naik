### Assignment 3: Set Operations

Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.

```python
set1 = set(range(1,6))
print("set 1: ",set1)

set2 = set(i for i in range(1,11) if i % 2==0)
print("set 2: ", set2)

## union operation:
set_union = set1.union(set2)
print(f"Union of sets: {set_union}")

## intersection of sets:
set_intersection = set1.intersection(set2)  
print(f"Intersection of Sets: {set_intersection}")

## set difference
set_difference = set1.difference(set2)
print(f"set difference: {set_difference}")

set_difference1 = set2.difference(set1)
print(f"set difference: {set_difference1}")

## symmetric difference
set_symmetric_diff = set1.symmetric_difference(set2)
print(f"Symmetric difference: {set_symmetric_diff}")

set_symmetric_diff1 = set2.symmetric_difference(set1)
print(f"Symmetric difference: {set_symmetric_diff1}")
```

```python
set1 = set(range(1, 6))
set2 = set(range(2, 11, 2))
print(f"Union: {set1 | set2}")
print(f"Intersection: {set1 & set2}")
print(f"Difference (set1 - set2): {set1 - set2}")
print(f"Symmetric Difference: {set1 ^ set2}")
```

```python
## Sets are unordered collection of data that means it has no index for elements
set_1 = {1,2,3,4,5}
try:
    set_1[1]
except TypeError:
    print("TypeError: 'set' object is not subscriptable i.e sets are unordered collections so no indexes")
```

### Assignment 7: Subsets and Supersets
